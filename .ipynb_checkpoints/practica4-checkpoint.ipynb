### 1. A partir de una imagen en escala de grises generar 10 im√°genes con ruido gaussiano. 

```python
#Imports
#******************************************************************************
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%pylab inline
```

```python
def imnoise(im, sigma, mu):
    
    M, N = im.shape
    R = (sigma * np.random.rand(M,N) + mu) + im
    
    return R
```

```python
path1 = 'Practica1/Imagenes/baboon.jpg'
path2 = 'Practica1/Imagenes/Globos.bmp'
img1 = cv2.imread(path1, 0)
img2 = cv2.imread(path2, 0)
```

```python
#Genero las imagenes con ruido y las ploteo automaticamente
for x in range(1,10):
    globals()['R' + str(x)] = imnoise(img1, 100 * x, 10 + x)
fig = plt.figure(figsize=(10,10))
for i in range(1,10):
    ax = plt.subplot(3,3,i)
    ax.imshow(globals()['R' + str(i)],cmap='gray')
    
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
#Genero las imagenes con ruido y las ploteo automaticamente
for x in range(1,10):
    globals()['R' + str(x)] = imnoise(img2, 100 * x, 10 + x)
fig = plt.figure(figsize=(10,10))
for i in range(1,10):
    ax = plt.subplot(3,3,i)
    ax.imshow(globals()['R' + str(i)],cmap='gray')
    
    ax.set_xticks([])
    ax.set_yticks([])
```
