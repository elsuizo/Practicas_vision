###Mostrar los efectos del muestreo y la cuantización en una imagen dada.

```julia
# Imports
import Images
import PyPlot
```

```julia
path1 = "../Imagenes/Globos.bmp";
img = Images.imread(path1);
img_gray = Images.rgb2gray(img);
```

```julia
PyPlot.imshow(img_gray,cmap="gray")
PyPlot.title("Imagen Original(escala de grises)")
```

```julia
# Primero prealocamos con zeros y luego le pasamos con slicing los puntos 

img_sub = zeros(size(img_gray))
img_sub[1:4:end, 1:4:end] = img_gray[1:4:end, 1:4:end];
```

```julia
PyPlot.imshow(img_sub,cmap="gray")
PyPlot.title("Imagen (escala de grises)")
```
