````julia
function histogram3(img)
    vals = linspace(FixedPointNumbers.Ufixed8(0),FixedPointNumbers.Ufixed8(1),256)
    h = Array(Float64,256)
    m,n = size(img)
    N = m * n
    for l = 1:length(vals)
        counter = 0.0
        @nloops 2 i img begin
            if((@nref 2 img i) == vals[l])
                counter += 1
            end
            h[l] = counter
        end
    end
    return h ./ N
end
````
