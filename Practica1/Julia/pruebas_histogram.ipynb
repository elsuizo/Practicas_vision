```julia
function histogram(img)
    vals = linspace(Ufixed8(0),Ufixed8(1),256)
    h = Array(Float64,256) # prealocate a Int array
    m,n = size(img)
    num_pix = m*n
    for i = 1:length(vals)
        h[i] = length(find(img .== vals[i]))
    end
    return h ./ num_pix # normalization
end
```

```julia
function histogram2(img)
    vals = linspace(Ufixed8(0),Ufixed8(1),256)
    h = Array(Float64,256)
    m,n = size(img)
    N = m * n
    for l = 1:length(vals)
        count = 0.0
        for i = 1:m
            for j = 1:n
                if(img[i,j] == vals[l])
                    count += 1
                end
                h[l] = count
            end
        end
    end
    return h ./ N
end

      
```

```julia
function histogram3(img)
    vals = linspace(Ufixed8(0),Ufixed8(1),256)
    h = Array(Float64,256)
    m,n = size(img)
    N = m * n
    for l = 1:length(vals)
        counter = 0.0
        @nloops 2 i img begin
            if((@nref 2 img i) == vals[l])
                counter += 1
            end
            h[l] = counter
        end
    end
    return h ./ N
end
```
