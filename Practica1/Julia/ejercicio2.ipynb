```julia
# Imports 
import Images
import PyPlot
```

```julia
path4 = "../Imagenes/sci-4.bmp";
```

```julia
img4 = Images.imread(path4);
```

```julia
# Convert to Array
img4_arr = convert(Array, img4);
```

```julia
PyPlot.imshow(img4_arr)
```

```julia
img4_red = img4_arr[:,:,1];
```

```julia
row_37 = img4_red[37,:]
```

###Graficamos la fila 37 como vector y vemos que se puede observar mucha dispersión dado que al haber poca iluminación el sensor CCD amplifica el ruido.
